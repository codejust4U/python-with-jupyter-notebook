### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
import math
x1 = int(input("enter the value of x1: "))
x2 = int(input("enter the value of x2: "))
y1 = int(input("enter the value of y1: "))
y2 = int(input("enter the value of y2: "))
ed = math.sqrt((x2-x1)**2+(y2-y1)**2)
print("the equlidean distance between two points i.e x1,y1 and x2,y2 is : ",ed)
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
numerator1 = int(input("enter the numerator 1: "))
numerator2 = int(input("enter the numerator 2: "))

denominator1 = int(input("enter the denominator 1: "))
denominator2 = int(input("enter the denominator 2: "))

total = ((numerator1*denominator2)/(denominator1*denominator2))+((numerator2*denominator1)/(denominator2*denominator1))
print("the sum of two fraction in decimal :",total)
num=total*(denominator1*denominator2)
deno=denominator1*denominator2
print("The sum of two fractions in fraction form :",num,"/",deno)
````
