### Gradient Descent


- gradient descent (also often called steepest descent) is a first-order iterative optimization algorithm for finding a local minimum of a differentiable function. 


```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

### applying OLS

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='blue')
```

### applying gradient descent assuming slope = 78.35 and b=0

```python
y_pred = ((78.35*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='blue',label='OLS')
plt.plot(X,y_pred,color='red',label='b=0')
plt.legend
```

```python
m=78.35
b=0
loss_slope = -2 * np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
# taking learning rate = 0.1
lr=0.1

step_size=loss_slope*lr
step_size
```

```python
b=b-step_size
b
```

```python
y_pred1= ((78.35*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='blue',label='OLS')
plt.plot(X,y_pred1,color='red',label='b={}'.format(b))
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
```

```python
X,y = make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20,random_state=13)
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
np.mean(cross_val_score(lr,X,y,scoring='r2',cv=10))
```

```python
class GDRegressor:
    def __init__(self,learning_rate,epochs):
        self.m = 29.19
        self.b=-120
        self.lr=learning_rate
        self.epochs=epochs

    def fit(self,X,y):
        # calculate the b usig GD
        for i in range(self.epochs):
            loss_slope=-2*np.sum(y-self.m*X.ravel()-self.b)
            self.b=self.b-(self.lr*loss_slope)
            print(loss_slope,"\t",self.b)
        print(self.b)
```

```python
gd = GDRegressor(0.001,100)
```
