### Order of Growth

- most of the programmers prefer average case scenario to build, but we will prefer worst case scenario so that our model wont make any mistake as average case does.
- we will use nested for loop rather than for loop
- loop after big order of operations

```python
def fact_iter(n):
    ans = 1
    while n>1:
        ans *= n
        n -= 1
    return ans

fact_iter(5)
```

```python
def i_to_str(i):
    digits = '0123456789'
    if i==0:
        return "0"
    result = ''
    while i>0:
        result=digits[i%10]+result
        i=i//10

    return result

i_to_str(4356)

# in this problem, as i have 4356 so it will run 4 times and incresing a digit in original makes it ten times of that digit, which affrec increase in +1 of time, which means its logarithmic 
```

### Q7

```python
def fibonacci(n):
    if n==1 or n==0:
        return 1
    else:
        return fibonacci(n-1)+fibonacci(n-2)
    
fibonacci(5)

# here the time complexity is exponential as input is been added and output is incresing with multiple which is inverse of logarithmic.
```

```python
class Solution:
    # Function to delete middle element of a stack.
    def deleteMid(self, s, sizeOfStack):
        first = 0
        last = len(s) - 1
        mid = (first + last) // 2
        s_list = list(s)
        for i,values in enumerate(s_list):
            if i == mid:
                s_list.pop(i)
                break  # Stop the loop after deleting the middle element
        print(s_list)

s = {10, 20, 30, 40, 50}
ob = Solution()
ob.deleteMid(s, len(s))
```
