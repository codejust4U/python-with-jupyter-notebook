### SibSp and Parch + 1 = Family_size

```python
dataset['SibSp']+dataset['Parch'] + 1
dataset['Family_size'] = dataset['SibSp']+dataset['Parch'] + 1
dataset
```

### family type
- 1> alone
- 2-4>small
- 5>large

```python
def transform_family_size(num):
    if num==1:
        return 'alone'
    elif num>1 and num<5:
        return 'small'
    else:
        return 'large'
```

```python
dataset['Family_type']=dataset['Family_size'].apply(transform_family_size)
```

```python
dataset
```

```python
pd.crosstab(dataset['Survived'],dataset['Family_type'],normalize='columns')*100
```
