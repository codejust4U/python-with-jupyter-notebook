# Calulations for the Backward Elimination

```python
# importing the libraries
import pandas as pd

# reading the data
df = pd.read_excel(r'D:\datas\excel\Day14\datatab.xlsx')

# initialization
n = 40
k = 5
ss_error = 93.5
ss_total = 150.3

# calculation
Total = n-1
Residual = n-k-1
ssr = ss_total - ss_error
MSR = ssr/k
MSE = ss_error - (n-k-1)
F = MSR/MSE
R_square = ssr/ss_total


plot_ = pd.DataFrame({"DF_value":[k,Residual,Total,R_square],"SS_value":[ssr,ss_error,ss_total,'null'],"MS_value":[MSR,MSE,'null','null'],"F_value":[F,'null','null','null']})

plot_
```

```python
import numpy as np
Adjusted_R_sq = 1-((Total/Residual)*(1-R_square))
print("The Adjusted R Squared is: ", Adjusted_R_sq)

CSE = np.sqrt(ss_error/Residual)
print("The computed standard error is: ",CSE)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel(r'D:\datas\excel\Day14\datatab.xlsx')

def gradient_descent(m_now, b_now, points, L):
    m_gradient = 0
    b_gradient = 0
    n = len(points)

    for i in range(n):
        x = points.iloc[i]['Salary']
        y = points.iloc[i]['Age']

        m_gradient += -(2/n)*x*(y - (m_now*x + b_now))
        b_gradient += -(2/n)*(y - (m_now*x + b_now))

    m = m_now - m_gradient * L
    b = b_now - b_gradient * L

    return m, b

m = 0
b = 0
L = 0.0001
epochs = 1000

for i in range(epochs):
    m, b = gradient_descent(m, b, data, L)

print(m, b)
f = [1250,3100]
g = [22.5,50]
plt.scatter(data['Salary'], data['Age'])
plt.plot(f,g)
#plt.plot(range(10, 60), [m*x + b for x in range(10, 60)])
plt.show()
```
