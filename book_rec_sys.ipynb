```python
x=ratings['user_id'].value_counts()>200
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings_with_books = ratings.merge(books,on='ISBN')
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
number_rating=ratings_with_books.groupby('title')['book_rating'].count().reset_index()
```

```python
number_rating.rename(columns={'book_rating':"no_of_rating"},inplace=True)
```

```python
final_rating=ratings_with_books.merge(number_rating,on='title')
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating=final_rating[final_rating['no_of_rating']>=50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user_id',index='title',values='book_rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
np.where(book_pivot.index==237)
```

```python
distances,suggestions = model.kneighbors(book_pivot.iloc[237, :].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index==book_name)[0][0]
    distances,suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1,-1),n_neighbors=6)

    for i in range(len(suggestions)):
            if i==0:
                 print("The suggestions for",book_name,"are :")
            if not i:
                print(book_pivot.index[suggestions[i]])
```

```python
recommend_book('Harry Potter and the Chamber of Secrets (Book 2)')
```
