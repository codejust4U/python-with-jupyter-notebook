### 1. Relational Plot
- Statistical analysis is a process of understanding how variables in a dataset relate to each other and how those relationships depend on other variables.
- The one we will use most is relplot(). This is a figure-level function for visualizing statistical relationships using two common approaches: scatter plots and line plots. relplot() combines a FacetGrid with one of two axes-level functions:

scatterplot() (with kind="scatter"; the default)

lineplot() (with kind="line")

```python
tips = sns.load_dataset('tips')
tips
```

### Scatter plot

```python
# Scatter plot
# axis level plot
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
# using hue on above code and style , size
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
# figure level plot
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex')
```

### lineplot

```python
gap = px.data.gapminder()
gap
```

```python
temp_df=gap[gap['country']=='India']
temp_df
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
t_df = gap[gap['country'].isin(['India','Pakistan','China'])]
t_df
```

```python
# Figure Level
sns.relplot(data=t_df,x='year',y='lifeExp',kind='line',hue='country')
```

```python
# Axis Level
sns.lineplot(data=t_df,x='year',y='lifeExp',hue='country')
```

```python
t_df2 = gap[gap['country'].isin(['Nepal','Germany','Iraq','Brazil'])]
t_df2
```

```python
sns.relplot(kind='line',data=t_df2,x='year',y='lifeExp',hue='country',style='continent',size='continent')
```

### 2. Distribution Plot
- Techniques for distribution visualization can provide quick answers to many important questions. What range do the observations cover? What is their central tendency? Are they heavily skewed in one direction? Is there evidence for bimodality? Are there significant outliers? Do the answers to these questions vary across subsets defined by other variables?
- The distributions module contains several functions designed to answer questions such as these. The axes-level functions are histplot(), kdeplot(), ecdfplot(), and rugplot(). They are grouped together within the figure-level displot(), jointplot(), and pairplot() functions.

- Figure level -> displot
- Axes level -> histplot, kdeplot,rugplot

### Histplot

```python
# univaraite histogram
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist',bins=20)
```

```python
tips
```

```python
# acting like a countplot
sns.displot(data=tips,x='day',kind='hist')
```

```python
sns.displot(data=tips,x='tip',hue='sex',kind='hist')
```

```python
sns.displot(data=tips,x='tip',hue='sex',kind='hist',element='step')
```

### 3. Matrix plot

```python
t_df3 = gap.pivot(index='country',columns='year',values='lifeExp')
```

### heatmap

```python
# axes level function
plt.figure(figsize=(15,15))
sns.heatmap(t_df3)
```

```python
# for one continent
t_df4=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
```

```python
# axes level function
plt.figure(figsize=(15,15))
sns.heatmap(t_df4,annot=True,linewidths=0.5,cmap='viridis')
```
